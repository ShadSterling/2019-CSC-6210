```python
# from curtsies import Input  ## error: (25, 'Inappropriate ioctl for device')
import time
import easygopigo3
import traceback

robot = easygopigo3.EasyGoPiGo3()
servo = robot.init_servo()
distance = robot.init_distance_sensor()

def doContinue():
    time.sleep( 0.2 )

def doStop():
    robot.stop()

def goForward():
    robot.forward()

def goBackward():
    robot.backward()

def turnLeft():
    robot.left()

def turnRight():
    robot.right()

def getRange(): return distance.read_mm()

servo.rotate_servo(90) ## set distance servo to middle position (forward)
done = False
while not done:
    for cmd in input().lower():
#     with Input(keynames = "curtsies") as keyboard:  ## error: (25, 'Inappropriate ioctl for device')
#         cmd = keyboard.send()
        
        try:
            if cmd == "q" or cmd == "x":
                print( cmd+" - Done, will exit at end of loop" )
                done = True
            elif cmd == "i":
                print( cmd+" - Moving Forward" )
                goForward()
            elif cmd == "j":
                print( cmd+" - Turning Left" )
                turnLeft()
            elif cmd == "k":
                print( cmd+" - Moving Backward" )
                goBackward()
            elif cmd == "l":
                print( cmd+" - Turning Right" )
                turnRight()
            elif cmd == "n" or cmd == "3":
                print( cmd+" - Turning North" )
                turnNorth()
            elif cmd == "w":
                print( cmd+" - Turning West" )
                turnWest()
            elif cmd == "e":
                print( cmd+" - Moving East" )
                turnEast()
            elif cmd == "s":
                print( cmd+" - Turning South" )
                turnSouth()
            elif cmd == ",":
                print( cmd+" - Delay next command" )
                doContinue()
            elif cmd == " ":
                print( cmd+" - Stop Moving" )
                doStop()
            else:
                print( cmd+" - Invalid command!")
            time.sleep( 0.2 )
#             print( "Heading: "+getHeading()+", Range: "+getRange() )
            print( "Range: "+str(getRange()) )
        except Exception as error:
            print( "!!! FAULT: ", traceback.format_exc() )
print( "Exit loop ended, exiting now" )
robot.stop()
```
